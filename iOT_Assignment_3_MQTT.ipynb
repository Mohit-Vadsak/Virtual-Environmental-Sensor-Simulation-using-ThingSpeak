```python
import paho.mqtt.client as mqtt
import random
import time

# Unique ID for the virtual station
station_id = "virtual_station01"

# ThingSpeak Channel settings
channel_id = "2488583"  # Replace with your actual Channel ID from ThingSpeak
write_api_key = "XQ69D9G7X001DHH1"  # Replace with your actual Write API Key from ThingSpeak

# MQTT settings
mqtt_host = "mqtt3.thingspeak.com"
mqtt_user = "NRwLCggvCSY7BCc0LysbLz0"
mqtt_password = "CmqcRqiDjLhxEB8LWTR23aic"
mqtt_topic = f"channels/{channel_id}/publish"

# --- Main Code ---



def generate_sensor_data():
    """Simulate generating random sensor values."""
    temperature = random.uniform(-50, 50)  # Temperature range: -50 to 50 Celsius
    humidity = random.uniform(0, 100)       # Humidity range: 0 to 100%
    co2 = random.uniform(300, 2000)         # CO2 range: 300ppm to 2000ppm
    return temperature, humidity, co2

def on_connect(client, userdata, flags, rc):
    """Callback function for MQTT client connection."""
    if rc == 0:
        print("Connected to MQTT broker")
    else:
        print("Connection failed")

def on_publish(client, userdata, mid):
    """Callback function for when a message is published."""
    print(f"Message published, MID: {mid}")

def publish_sensor_data(temperature, humidity, co2):
    """Publish the sensor data to ThingSpeak channel via MQTT."""
    payload = f"field1={temperature}&field2={humidity}&field3={co2}"
    client.publish(mqtt_topic, payload)

# Create MQTT client (with callback API Version)
client = mqtt.Client(callback_api_version=mqtt.CallbackAPIVersion.VERSION1, client_id=mqtt_user, protocol=mqtt.MQTTv311)
client.username_pw_set(username=mqtt_user, password=mqtt_password)

# Set up MQTT client callbacks
client.on_connect = on_connect
client.on_publish = on_publish  # Assign the on_publish callback

# Connect to MQTT broker
client.connect(mqtt_host, 1883)


# Main loop
try:
    while True:
        temp, hum, co2_level = generate_sensor_data()
        print(f"Generated data - Temperature: {temp:.2f} C, Humidity: {hum:.2f}%, CO2: {co2_level:.2f} ppm")
        publish_sensor_data(temp, hum, co2_level)
        time.sleep(20)  # Respect ThingSpeak update limits
except KeyboardInterrupt:
    print("Simulation stopped by user.")
    client.disconnect()
```
